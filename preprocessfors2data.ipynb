```python
# Import libraries
import numpy as np
import pandas as pd
import scipy.io
import os
import json
```

```python
# Thư mục chứa các file .mat
dataset_path = "hand_dataset/validation_dataset/validation_data"
json_output_path = "hand_dataset"
mat_path = os.path.join(dataset_path, "annotations")

def read_data(boxes):
    try:
        k = len(boxes[0][0][0][0])
        data = []
        if k == 0:
            return data
        for i in range(k):
            try:
                value = []
                value.append(boxes[0][0][0][0][i][0][0])  # x1
                value.append(boxes[0][0][0][0][i][0][1])  # y1
                value.append(boxes[0][1][0][0][i][0][0])  # x2
                value.append(boxes[0][1][0][0][i][0][1])  # y2
                data.append(value)
            except:
                pass
        return data
    except:
        return []
    

def process_mat_file(mat_path):
    mat = scipy.io.loadmat(mat_path)
    boxes = mat['boxes']
    data = read_data(boxes)
    print(data)
    return data

def process_all_mat_files(mat_path):
    data_dict = {}
    for filename in os.listdir(mat_path):
        if filename.endswith(".mat"):
            filepath = os.path.join(mat_path, filename)
            data = process_mat_file(filepath)
            file_key = os.path.splitext(filename)[0]  # Loại bỏ đuôi .mat
            data_dict[file_key] = data
    return data_dict

def save_to_json(data_dict, output_file):
    with open(output_file, 'w') as json_file:
        json.dump(data_dict, json_file, indent=4)

# Tên file JSON đầu ra
output_file = 'validation_dataset.json'

# Xử lý tất cả các file .mat và lưu dữ liệu vào JSON
data_dict = process_all_mat_files(mat_path)
print(len(data_dict))
save_to_json(data_dict, os.path.join(json_output_path, output_file))

print(f"Data has been successfully saved to {output_file}")
```
